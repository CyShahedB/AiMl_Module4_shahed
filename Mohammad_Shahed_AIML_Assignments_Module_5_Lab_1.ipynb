#Changing degree in polynomial regression

```python
import numpy as np
from sklearn import linear_model,metrics
from sklearn.preprocessing import PolynomialFeatures
from matplotlib import pyplot as plt
```

```python
# POLYNOMIAL REGRESSION

def poly_regression(x,y,deg):

  x = x.reshape((-1,1))
  y = y.reshape((-1,1))


  poly = PolynomialFeatures(degree=deg)

  x_ = poly.fit_transform(x)
  poly.fit(x_,y)

  m1 = linear_model.LinearRegression()
  m1.fit(x_,y)
  y_pred = m1.predict(x_)

  mse = np.sqrt(metrics.mean_squared_error(y_pred,y))

  plt.plot(x,y_pred,color="red")
  plt.title("Visualization of actual data vs predicted outputs (polynomial regression)")
  plt.scatter(x,y,color="blue")
  plt.legend(["predicted polynomial","actual data points"])
  plt.show()

  print("The RMSE for polynomial regression is: {}".format(mse))
```

```python
deg = 24#@param {type:"slider", min: 1, max:30, step:1}

np.random.seed(10)

x_ex = np.array([i*np.pi/180 for i in range(60,300,6)])
y_ex = np.cos(x_ex) + np.random.normal(0,0.15,len(x_ex))
print(x_ex)
print(y_ex)
poly_regression(x_ex,y_ex,deg)
plt.show()
```

Here I took a dataset where x_ex is input datapoints values and y_ex is output datapoints.


Here x_ex has angles in radious. These angle values are starts from 60 degree to 300 degree and takes 6 step of degree at a time and then they converted into radions.


Here y_ex has cosine values of x_es values anong with some random values.


Now we are performing polynomial regression on it with various degree.


If we take low degree, the regression line is not fitting with the datapoints or data samples properly. This is called underfitting. This is done usually when machine learning model not trained properly. It leads poor predictions with test data.


If we take high degree, the regression line is trying to fit maximum datapoints or data samples also including unnecessary and noisy data. This is called overfitting. This is done usually when machine learning model trained morely on all data including noisy also. As a result, an overfitted model may perform very well on the training data but poorly on new, unseen data because it essentially memorizes the training set instead of generalizing to new patterns.
